# Variando $D$ y $\epsilon$

De los parámetros iniciales, vamos a varian unicamnete estos dos debido a que son complicados para medir, lo que hace que tengan un gran error. 

Nuestra idea es, realizar varias simulaciones variando estos parámetros, calcular la distancia cuadratica entre los valos obtenidos y los valores experimentales del paper de Mc Neal y ver como varia esta distancia.

```python
import os
import sys
sys.path.append('../')

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from matplotlib import colorbar
%matplotlib inline

import ten
```

```python
def clasificar_por_e(data_TPP):
    e1 = []
    e2 = []
    e3 = []
    e4 = []
    e5 = []
    e6 = []

    for i in range(180):
        e = data_TPP[i, 1]
        if e == 0.5:
            e1.append([data_TPP[i, 0], data_TPP[i, 3]])
        elif e == 1.0:
            e2.append([data_TPP[i, 0], data_TPP[i, 3]])
        elif e == 1.5:
            e3.append([data_TPP[i, 0], data_TPP[i, 3]])
        elif e == 2.0:
            e4.append([data_TPP[i, 0], data_TPP[i, 3]])
        elif e == 2.5:
            e5.append([data_TPP[i, 0], data_TPP[i, 3]])
        elif e == 3.0:
            e6.append([data_TPP[i, 0], data_TPP[i, 3]])
    return [e1, e2, e3, e4, e5, e6]
            
def ordenar_e(e1):
    e1_ordenada = []
    for i in range(len(e1)):
        minimo = [10000, 10000]
        for element in e1:
            if len(e1_ordenada) == 0:
                if element[0] < minimo[0]:
                    minimo = element
            else:
                if element[0] < minimo[0] and element not in e1_ordenada:
                    minimo = element
        e1_ordenada.append(minimo)
    return e1_ordenada

def ordenar_epsilons(epsilons):
    nuevo_epsilons = []
    for i in range(len(epsilons)):
        nuevo_e = ordenar_e(epsilons[i])
        nuevo_epsilons.append(nuevo_e)
    return nuevo_epsilons
```

Datos del paper de Mc. Neal. 

```python
# Datos del paper de McNeal

# Perylene:
aceptor_perylene = [0, 21, 43, 86, 171, 436, 871]
eff_perylene = [0, 0.1875, 0.3125, 0.68, 0.78, 0.9, 0.95]

# Coumarin 6
aceptor_coumarin = [0, 20, 60, 120, 240, 600, 1000]
eff_coumarin = [0, 0.42, 0.58, 0.85, 0.92, 0.97, 0.98]

# TPP
aceptor_TPP = [0, 40, 100, 200, 400, 1000]
eff_TPP = [0, 0.48, 0.82, 0.9, 0.95, 0.98]
```

```python
# Perilene
data_per = ten.post.extrac4dir('data/perilene/', aceptor_perylene)
data_per = ten.post.diference2(eff_perylene, data_per)

# Coumarine
data_cou = ten.post.extrac4dir('data/coumarin/', aceptor_coumarin)
data_cou = ten.post.diference2(eff_coumarin, data_cou)

# TPP 
data_TPP = ten.post.extrac4dir('data/TPP/', aceptor_TPP)
data_TPP = ten.post.diference2(eff_TPP, data_TPP)

e_tpp = ordenar_epsilons(clasificar_por_e(data_TPP))
e_cou = ordenar_epsilons(clasificar_por_e(data_cou))
e_per = ordenar_epsilons(clasificar_por_e(data_per))
```

```python
D = list(range(1, 5, 1)) + list(range(5, 61, 5)) + list(range(70, 101, 10)) + list(range(150, 601, 50))
epsilon = list(np.arange(0.5, 3.5, 0.5))

D_f = D * 6
epsilon_f = [0.5] * 30 + [1] * 30 + [1.5] * 30 + [2] * 30 + [2.5] * 30 + [3] * 30
```

```python
z = np.zeros((6, 30))
for i in range(6):
    for j in range(30):
        z[i, j] = e_tpp[i][j][1] / 6 + e_cou[i][j][1] / 7 + e_per[i][j][1] / 7
        
z_new = []
for i in range(len(z)):
    z_new.append(list(z[i]))
z_f = []
for i in z_new:
    for j in i:
        z_f.append(j)
```

```python
fig = plt.figure(figsize=(12, 12), dpi=300)
ax = fig.gca(projection='3d')
surf = ax.plot_trisurf(D_f, epsilon_f, z_f, cmap=cm.jet, linewidth=0.02)
ax.set_xlabel(r'$Mean\; Free\; Path\; (D)$', fontsize=20)
ax.set_ylabel(r'$\epsilon$', fontsize=20)
ax.set_zlabel(r'$Diferencia\; cuadrada$', fontsize=20)

cbar = plt.colorbar(surf, orientation='horizontal', shrink=0.8)
plt.savefig('Coumarine.svg', transparent=True)

l, b, w, h = plt.gca().get_position().bounds
cbar.ax.set_position([l, b + 0.1*h, w, h*0.95])
plt.show()
```

Podemos observar que tomando $\epsilon$ más chicos (entere 0.1 y 0.5nm) es donde se minimiza la distancia cuadratica, pero tambien es donde el tiempo de la simulación crece en forma exponencial (eso lo pedemos ver [acá](http://nbviewer.ipython.org/github/pewen/ten/blob/master/IPython_notebook/07_Profile_serial.ipynb#Variando-$\epsilon$-y-$D$)). Eso se debe a la relación entre los parámetros:

\begin{equation}
N = \dfrac{D}{\epsilon}
\end{equation}
