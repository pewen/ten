```python
# Datos del paper de McNeal

# Perylene:
aceptor_perylene = [0, 21, 43, 86, 171, 436, 871]
eff_perylene = [0, 0.1875, 0.3125, 0.68, 0.78, 0.9, 0.95]

# Columarin 6
aceptor_coumarin = [0, 20, 60, 120, 240, 600, 1000]
eff_coumarin = [0, 0.42, 0.58, 0.85, 0.92, 0.97, 0.98]

# TPP
aceptor_TPP = [0, 40, 100, 200, 400, 1000]
eff_TPP = [0, 0.48, 0.82, 0.9, 0.95, 0.98]
```

```python
r = 15 #coment
R_Forster = 2.29
L_D_init = 100
L_D_end = 301
tau_D = 0.333
epsilon_init = 1
epsilon_end = 3
num_exc = 5000

num_acceptors_min = 0
num_acceptors_max = 437
acceptors_step = 1

for epsilon in np.linspace(1, 3, 5):
   for L_D in range(L_D_init, L_D_end, 50):
        # Inicialice some variables
        output_parameters = []

        for num_acceptors in range(num_acceptors_min,
                                   num_acceptors_max,
                                   acceptors_step):
            num_acceptors = int(num_acceptors)

            # Initialice the nanopartile object,

            nano_particle = NanoParticle(r,
                                         num_acceptors,
                                         tau_D,
                                         R_Forster,
                                         L_D,
                                         epsilon,
                                         'vol')

            # Initialece the exiton,
            simu = Exciton(nano_particle,
                           num_exc,
                           'laser')

            #Calculate
            simu.quenching(each=True)
            output_parameters += [list(simu.get_output())]

        input_parameters = simu.get_input()
        input_parameters.append([x for x in range(num_acceptors_min,
                                                  num_acceptors_max,
                                                  acceptors_step)])

        save_out(input_parameters, output_parameters, file_path='perilene/')

        print('se termino la simulacion con epsilon = %f, L_D = %f' %(epsilon, L_D))
```

```python
def get_efficiency(dir_path, search):
    """
    Extrae la eficiencia de todos los archivos dentro de un directorio y los retorna en una matriz.
    
    La primer columna es el mean free path.
    La segunda columna es el epsilon.
    Las restantes son las eficienecias.
    
    Parameters
    ----------
    dir_path: str
        Path al directorio donde se va a buscar los archivos
    search: list
        List con los valores de aceptores para los cuales se quiere extraer la eficiencia.
    """
    dirs = os.listdir(path=dir_path)

    #search = [0, 40, 100, 200, 400]
    out = np.zeros((len(dirs), len(search) + 2), dtype=np.float64)

    for num_file, file_path in enumerate(dirs):
        with open(os.path.join(dir_path, file_path), 'r') as file:

            cnt = 2

            while True:
                line = file.readline()

                if 'path' in line:
                    line_split = line.split()
                    out[num_file][0] = float(line_split[3])

                elif 'Epsilon' in line:
                    line_split = line.split()
                    out[num_file][1] = float(line_split[1])

                elif 'Nº acceptors' in line:
                    line = file.readline()
                    while True:
                        line = file.readline()

                        #Remove all spaces
                        line_without_space = ''.join(line.split())
                        line_split = line_without_space.split('|')

                        if '+--------------' in line:
                            break

                        if line_split[1] == str(search[cnt - 2]):
                            out[num_file][cnt] = float(line_split[4])
                            cnt += 1

                    break

                if '' == line:
                    break
                    
    return out



def diff_efficiency(out, eff_compare):
    """
    
    """
    len_dir, len_search = out.shape
    len_search -= 2
    
    error = np.zeros((len_dir, 3))
    error[:, :2] = out[:, :2]

    eff = np.zeros((len_dir, len_search))
    eff = out[:, 2:]

    dif_square = (eff - eff_compare)**2
    error[:, -1] = dif_square[:, 0] + dif_square[:, 1] + dif_square[:, 2]+ dif_square[:, 3] + dif_square[:, 4]
    
    return error
```

```python
out_TPP = get_efficiency('TPP/', aceptor_TPP)

difference_TPP = diff_efficiency(out_TPP, eff_TPP)
print(difference_TPP)
```

```python
out_coumarin = get_efficiency('coumarin/', aceptor_coumarin)

difference_coumarin = diff_efficiency(out_coumarin, eff_coumarin)
print(difference_coumarin)
```

```python
out_perylene = get_efficiency('perilene/', aceptor_perylene)

difference_perylene = diff_efficiency(out_perylene, eff_perylene)
print(difference_perylene)
```
