#SpeedUp

Comparando el tiempo del código serial con el paralelo

```python
# Imports necesarios

import time

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Archivo de configuración para medir los tiempos
# Solo se varia la cantidad de excitaciones

def experient_conf(num_exc):
    conf = """r_mean = 15 #coment
R_Forster = 3.14
mean_path = 100
tau_D = 0.333
epsilon = 1.5
    
num_acceptors_min = 0
num_acceptors_max = 501
acceptors_step = 20
num_exc = %i

acceptors = vol
exiton = laser
r_electro = 0""" %num_exc
    
    conf_file = open('speed.conf', 'w')
    conf_file.write(conf)
    conf_file.close
```

```python
# Medición de los tiempos

serial_time = []
parallel2 = []
parallel4 = []

x = np.array([i for i in range(1000, 10001, 1000)])

for num_exc in x:
    experient_conf(num_exc)
    
    start = time.time()
    !python3 ../main.py -c speed.conf -o ../output/
    serial_time.append(time.time() - start)
    
    start = time.time()
    !mpirun -n 2 python3 ../main.py -c speed.conf -o ../output/
    parallel2.append(time.time() - start)
    
    start = time.time()
    !mpirun -n 4 python3 ../main.py -c speed.conf -o ../output/
    parallel4.append(time.time() - start)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(x, serial_time, '-o', linewidth=2, label="Tiempo serial")
plt.plot(x, parallel2, 'r-*', linewidth=2, label="Tiempo paralelo 2 pros")
plt.plot(x, parallel4, 'g--', linewidth=2, label="Tiempo paralelo 4 pros")
plt.xlabel(r'$Número\; de\; exitaciones$', fontsize=20)
plt.ylabel(r'$Tiempo(seg)$', fontsize=20)
plt.title(r'$SpeedUp$', fontsize=20)
#plt.text(1100, 230, r'$Ajuste\; ax+ b$', fontsize=20)
#plt.text(1100, 200, r'$a = %f b = %f$' %(popt[0], popt[1]), fontsize=20)
plt.legend(loc='best', fontsize=16)
plt.grid(True)
```
